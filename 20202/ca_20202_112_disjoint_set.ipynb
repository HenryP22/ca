```
import graphviz
def drawUF(a):
    dot = graphviz.Digraph(comment='Nada')
    n = len(a)
    for i in range(n):
        dot.node(str(i), str(i))
    for i in range(n):
        if a[i] > 0:
            dot.edge(str(i), str(a[i]))
    dot.graph_attr['rankdir'] = 'BT'
    return dot
```

# Union Find

## Quick Union Ponderado

```
def find(id, a):
    if id[a] > 0:
        rep = find(id, id[a])
        id[a] = rep
        return rep
    else:
        return a

def union(id, a, b):
    parentA = find(id, a)
    parentB = find(id, b)
    if parentA == parentB:
        return

    if -id[parentA] < -id[parentB]:
        id[parentB] += id[parentA]
        id[parentA] = parentB
    else:
        id[parentA] += id[parentB]
        id[parentB] = parentA
```
