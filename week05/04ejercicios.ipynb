````
import random as rnd
import graphviz as gv

def printG(G):
    for line in G:
        print(line)

def drawDirectedAdjMatrix(G):
    dot = gv.Digraph(comment='Nada')
    n = len(G)
    for u in range(n):
        dot.node(str(u), str(u))
    
    for u in range(n):
        for v in range(n):
            if G[u][v] == 1:
                dot.edge(str(u), str(v))

    dot.graph_attr['rankdir'] = 'BT'
    return dot


def drawDirectedAdjList(G):
    dot = gv.Digraph(comment='Nada')
    n = len(G)
    for u in range(n):
        dot.node(str(u), str(u))
    
    for u in range(n):
        for v in G[u]:
            dot.edge(str(u), str(v))

    dot.graph_attr['rankdir'] = 'BT'
    return dot
````

## Ejercicio 1: Creación de un grafo aleatorio G de n vértices y m aristas

 * Escriba un algoritmo que genere una matriz aleatoria de adyacencia para G

````
def randAdjMatrix(n, m):
    G = [[0]*n for _ in range(n)]

    cont = 0
    while cont < m:
        i = rnd.randint(0, n-1)
        j = rnd.randint(0, n-1)
        if G[i][j] == 0 and G[j][i] == 0 and i != j:
            G[i][j] = 1
            cont += 1

    return G
````

````
G = randAdjMatrix(5, 6)
````

````
printG(G)
````

````
drawDirectedAdjMatrix(G)
````

 * Escriba un algoritmo que genere una lista aleatoria de adyacencia para G

````
def randAdjList(n, m):
    G = [[] for _ in range(n)]

    cont = 0
    while cont < m:
        i = rnd.randint(0, n-1)
        j = rnd.randint(0, n-1)
        if not(j in G[i] or i in G[j] or i == j):
            G[i].append(j)
            cont += 1

    return G
````

````
G = randAdjList(6, 8)
````

````
printG(G)
````

Explicación:
```
[]          -> no arcos parten de vertice 0
[2, 0]      -> arcos: (1->2), (1->0)
[4, 5, 3]   -> arcos: (2->4), (2->5), (2->3)
[4]         -> arcos: (3->4)
[1]         -> arcos: (4->1)
[4]         -> arcos: (5->4)
```

````
drawDirectedAdjList(G)
````
