```
def bfsAdjList(G, s):
    n = len(G)
    visited = [False]*n
    parent = [None]*n
    queue = [s]
    visited[s] = True
    while len(queue) > 0:
        u = queue[0]
        queue = queue[1:]
        for v in G[u]:
            if not visited[v]:
                parent[v] = u
                visited[v] = True
                queue.append(v)

    return parent
```

```
G = [[4],
     [5, 6],
     [4],
     [7],
     [0, 2, 6, 7],
     [1, 6, 7],
     [1, 4, 5],
     [3, 4, 5]]
```

```
path = bfsAdjList(G, 4)
print(path)
```

```
import graphviz as gv

def drawAL(G, path=None):
    dot = gv.Graph(comment='Nada')
    n = len(G)
    for u in range(n):
        dot.node(str(u), str(u))
    
    for u in range(n):
        for v in G[u]:
            color = 'blue' if path == None else 'red'
            if path != None and (path[u] == v or path[v] == u):
                color = 'red'
            else:
                color = 'blue'
            dot.edge(str(u), str(v), color=color)

    dot.graph_attr['rankdir'] = 'BT'
    return dot
```

```
drawAL(G)
```

```
drawAL(G, path)
```
